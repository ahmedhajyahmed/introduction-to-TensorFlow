# Import libraries

```python
import numpy as np
# ploting graphs
import matplotlib.pyplot as plt
# nice 3D graphs
from mpl_toolkits.mplot3d import Axes3D
```

# Generate random data

```python
# inputs
observations = 1000
xs = np.random.uniform(low=-10,high=10,size=(observations,1))
zs = np.random.uniform(low=-10,high=10,size=(observations,1))
inputs = np.column_stack((xs,zs))
print(inputs.shape)
```

#### Element of the model in supervized learning : 
1) inputs
2) weights
3) biases
4) outputs
5) targets

```python
# target 
noise = np.random.uniform(low= -1, high = 1, size=(observations,1))
targets = 2*xs - 3*zs + 5 + noise
targets.shape
```

###  plotting the training data

```python
# in order to use the 3D plot, the objects should have a certain shape
targets = targets.reshape(observations,)

# plotting according to the pyplot syntax

# declare the figure 
fig = plt.figure()

# a methode allowing us to create a 3D plot
ax = fig.add_subplot(111, projection='3d')

# choose the axes
ax.plot(xs,zs,targets)

# set labels
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('targets')

# plot the data with an angle
ax.view_init(azim=100)

# show the plot 
plt.show()

targets = targets.reshape(observations,1)
```

### initialize weights

```python
weights = np.random.uniform(-0.1,0.1,size=(2,1))
biases = np.random.uniform(-0.1,0.1, size=1)
print(weights, biases)
```

```python
# set learning rate
learning_rate = 0.02
```

 ### train the model 

```python
for i in range(100):
    output = np.dot(inputs, weights) + biases
    # delta is a (1000,1) matrix 
    delta = output - targets
    
    # loss is a scalar
    # w / by observations to make normalized 
    loss = np.sum(delta ** 2) / 2 / observations
    print(loss)
    
    delta_scal = delta / observations
    
    weights = weights - learning_rate * np.dot(inputs.T, delta_scal)
    biases = biases - learning_rate * np.sum(delta_scal)
    
    
    
```

### print the last weights and biases

```python
print(weights, biases)
```

### print last output vs target

```python
plt.plot(output, targets)
plt.xlabel('output')
plt.ylabel('target')
plt.show
```
